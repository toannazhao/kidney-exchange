# Graphs

## 3D with fixed parameters

```python
def plot_3d(data, param_dict, x, y, metric, color = True):
    """
    Plot a 3D scatterplot of x and y, fixing a dictionary of parameters to their given values, with a metric on the z-axis.
    
    Params:
        data: dataframe. Must include column x, y, and metric
        param_dict: dictionary of parameter columns and their corresponding values
        x: column to plot on the x-axis
        y: column to plot on the y-axis
        metric: column to plot on the z-axis
        color: if true (default), differentiates the color of the scatterpoints by the metric to depict depth
    """
    # filter data to fix all specified parameters to some value
    dat = data.loc[(data[list(param_dict)] == pd.Series(param_dict)).all(axis=1)]
    
    fig = plt.figure(figsize = (8, 8))
    ax = plt.axes(projection='3d')
    
    # capS and capH as the x and y, metric on the z axis
    scatter = ax.scatter3D(dat[x], dat[y], dat[metric], c=color*dat[metric])
    fig.colorbar(scatter, shrink=0.5)
    
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    ax.set_zlabel(metric)
    ax.view_init(20, -120)
    
    ax.set_title("Capacities for interviews vs " + metric)
    plt.show()

plot_3d(df, {'alphaS': 0.5, 'alphaH': 0.5, 'betaS': 0.5, 'betaH': 0.5}, 'capS', 'capH', 'blockSfull')
plot_3d(df, {'alphaS': 0.5, 'alphaH': 0.5, 'betaS': 0.5, 'betaH': 0.5}, 'capS', 'capH', 'unassignedS')
```

```python
def plot_capacities_3d(data, param_dict, metric, color = True):
    """
    Plot capS and capH as x and y, fixing a dictionary of parameters to their given values, with a metric on the z-axis.
    
    Params:
        data: dataframe. Must include column x, y, and metric
        param_dict: dictionary of parameter columns and their corresponding values
        metric: column to plot on the z-axis
        color: if true (default), differentiates the color of the scatterpoints by the metric to depict depth
    """
    # filter data to fix all specified parameters to some value
    dat = data.loc[(data[list(param_dict)] == pd.Series(param_dict)).all(axis=1)]
    
    fig = plt.figure(figsize = (8, 8))
    ax = plt.axes(projection='3d')
    
    # capS and capH as the x and y, metric on the z axis
    scatter = ax.scatter3D(dat['capS'], dat['capH'], dat[metric], c=color*dat[metric])
    fig.colorbar(scatter, shrink=0.5)
    
    ax.set_xlabel("capS")
    ax.set_ylabel("capH")
    ax.set_zlabel(metric)
    ax.view_init(20, -120)
    
    ax.set_title("Capacities for interviews vs " + metric)
    plt.show()

plot_capacities_3d(df, {'alphaS': 0.5, 'alphaH': 0.5, 'betaS': 0.5, 'betaH': 0.5}, 'blockSfull')
plot_capacities_3d(df, {'alphaS': 0.5, 'alphaH': 0.5, 'betaS': 0.5, 'betaH': 0.5}, 'unassignedS')
```

## Heat map

```python
def plot_heatmap(data, param_dict, x, y, metric):
    """
    Plot a heatmap of the metric on values of x and y. Returns the heatmap dataframe.
    """
    dat = data.loc[(data[list(param_dict)] == pd.Series(param_dict)).all(axis=1)].pivot(x, y, metric)
    
    plt.figure(figsize = (8, 5))
    ax = sns.heatmap(dat)
    ax.set_title("Heatmap of " + metric + " on capacities")
    plt.show()
    
    return(dat)

plot_heatmap(df, {'alphaS': 0.5, 'alphaH': 0.5, 'betaS': 0.5, 'betaH': 0.5}, 'capS', 'capH', 'blockSfull')
```

```python
def plot_heatmap_cap(data, param_dict, metric):
    """
    Plot a heatmap of the metric on values of capS and capH. Returns the heatmap dataframe.
    """
    dat = data.loc[(data[list(param_dict)] == pd.Series(param_dict)).all(axis=1)].pivot("capS", "capH", metric)
    
    plt.figure(figsize = (8, 5))
    ax = sns.heatmap(dat)
    ax.set_title("Heatmap of " + metric + " on capacities")
    plt.show()
    
    return(dat)

plot_heatmap_cap(df, {'alphaS': 0.5, 'alphaH': 0.5, 'betaS': 0.5, 'betaH': 0.5}, 'blockSfull')
```

## Plots where capH == capS

```python
def plot_agg_metric(data, param_dict, grouped, metric, agg='mean', missing_capS=False):
    """
    Plot the metric, grouped and aggregated by a specified variable and controlled for parameters.
    Filter to observations where capH equals capS. 
    
    Params:
        data: dataframe. Must include columns 'capS', 'capH', metric, and grouped.
        param_dict: dictionary of parameter columns and their corresponding values
        metric: column to plot and aggregate
        agg: aggregation function. Default is mean
        missing_capS: if True, the dataframe has capS always set to a particular number. 
        Assume capS always equals capH in this case. Else, need to filter for when capS == capH.
        compare_df: Dataframe to compare to. Default is none.
    """
    dat = data.loc[(data[list(param_dict)] == pd.Series(param_dict)).all(axis=1)]
    if missing_capS:
        dat.capS = dat.capH
    dat = dat[dat.capS == dat.capH]
    dat = dat.groupby(grouped).agg(agg)
    
    plt.figure(figsize = (8, 5))
    plt.plot(dat.index, dat[metric], label = metric, marker = 'o')
    
    plt.suptitle(agg.title() + " metric by " + grouped)
    plt.title("Observations where capS and capH are equivalent")
    plt.xlabel(grouped)
#     plt.xticks(dat.index, dat.index)
    plt.legend()
    
    return(dat)

# compare df and df2
plot_agg_metric(df, {'alphaS': 0.5, 'alphaH': 0.5, 'betaS': 0.5, 'betaH': 0.5}, 'capS', 
                ['avgRankInterviewS', 'blockSfull', 'avgRankInterviewH'])
plot_agg_metric(df2, {'alphaS': 0.5, 'alphaH': 0.5, 'betaS': 0.5, 'betaH': 0.5}, 'capS', 
                ['avgRankInterviewS', 'blockSfull', 'avgRankInterviewH'], 'mean', True)
```
