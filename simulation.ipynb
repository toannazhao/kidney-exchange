# Simulations

```python
from queue_simulation import Queue     
```

```python
q = Queue(arrival_rate=10, departure_rate=1, compatibility=matrix, values=np.ones((len(matrix), len(matrix))),
         ids = list(ke_nkr.id))

num_generate = 100000
q.generate_arrivals(num_generate)
q.generate_departures(num_generate)

num_matches_goal = 1000
start_time = datetime.now()

while len(q.matches) < num_matches_goal:
    q.next_greedy_match()
    print("another match! current pool length is", len(q.current))
    
end_time = datetime.now()
print("time for ", num_matches_goal, "matches: ", end_time - start_time)
```
